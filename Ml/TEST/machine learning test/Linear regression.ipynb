```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("alzheimers_disease_data.csv")
```

```python
df.head(3)
```

```python
x = df['Age'].values  # Using 'AP' as a single feature for simplicity
y = df['Forgetfulness'].values
```

```python
print(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
ml = LinearRegression()
```

```python
x_train = np.vstack((np.ones_like(x_train), x_train)).T
```

```python
x_test = np.vstack((np.ones_like(x_test), x_test)).T
```

```python
ml.fit(x_train,y_train)
```

```python
y_pred = ml.predict(x_test)
print(y_pred)
```

```python
results_df = pd.DataFrame({
    'Actual Values': y_test,
    'Predicted Values': y_pred,
    'Difference (Actual - Predicted)': y_test - y_pred
})
print(results_df)
```
